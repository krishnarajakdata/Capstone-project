```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans


data = {
    'Murders': [10, 20, 5, 15, 8, 12],
    'Kidnappings': [5, 8, 3, 10, 6, 7]
}
df = pd.DataFrame(data)


scaler = StandardScaler()
scaled_features = scaler.fit_transform(df)


kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(scaled_features)


df['Cluster'] = clusters


print(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x='Murders', y='Kidnappings', hue='Cluster', data=df)
plt.title('Crime Rate Clustering')
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans


data = {
    'District': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
    'Crime_Rate': [50, 20, 30, 80, 70, 10, 60, 90, 20, 40],
    'Population_Density': [1000, 800, 1200, 3000, 2500, 700, 1500, 4000, 850, 1100],
    'Poverty_Rate': [20, 15, 18, 35, 30, 10, 25, 40, 12, 22],
    'Unemployment_Rate': [7, 6, 5, 12, 10, 4, 8, 15, 5, 6]
}

df = pd.DataFrame(data)

features = df[['Crime_Rate', 'Population_Density', 'Poverty_Rate', 'Unemployment_Rate']]


scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)


kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_features)

cluster_map = {0: 'Peaceful Area', 1: 'Moderate Area', 2: 'Sensitive Area'}
df['Cluster'] = df['Cluster'].map(cluster_map)


sensitive_areas_df = df[df['Cluster'] == 'Sensitive Area']
moderate_areas_df = df[df['Cluster'] == 'Moderate Area']
peaceful_areas_df = df[df['Cluster'] == 'Peaceful Area']


print("Sensitive Areas:")
print(sensitive_areas_df)
print("\nModerate Areas:")
print(moderate_areas_df)
print("\nPeaceful Areas:")
print(peaceful_areas_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x='Population_Density', y='Crime_Rate', hue='Cluster', data=df)
plt.title('Crime Rate vs. Population Density')
plt.xlabel('Population Density')
plt.ylabel('Crime Rate')
plt.show()
```

```python
sns.scatterplot(x='Poverty_Rate', y='Crime_Rate', hue='Cluster', data=df)
plt.title('Crime Rate vs. Poverty Rate')
plt.xlabel('Poverty Rate')
plt.ylabel('Crime Rate')
plt.show()
```

```python
sns.scatterplot(x='Unemployment_Rate', y='Crime_Rate', hue='Cluster', data=df)
plt.title('Crime Rate vs. Unemployment Rate')
plt.xlabel('Unemployment Rate')
plt.ylabel('Crime Rate')
plt.show()
```
