```python
import pandas as pd


data = {
    'State/UT': ['State1', 'State2', 'State3', 'State4', 'State5', 'State6', 'State7', 'State8', 'State9', 'State10'],
    'Total_Crimes': [5000, 3000, 7000, 2000, 6000, 4000, 1000, 8000, 500, 9000],
    'Literacy_Rate': [85, 78, 92, 74, 88, 81, 69, 95, 76, 90]
}

df = pd.DataFrame(data)
print(df)
```

```python

print(df.describe())
```

```python

correlation = df['Literacy_Rate'].corr(df['Total_Crimes'])
print(f"Correlation between Literacy Rate and Total Crimes: {correlation}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(10, 6))
sns.scatterplot(x='Literacy_Rate', y='Total_Crimes', data=df)
plt.title('Literacy Rate vs Total Crimes')
plt.xlabel('Literacy Rate (%)')
plt.ylabel('Total Crimes')
plt.show()
```

```python

data = {
    'State/UT': ['State1', 'State2', 'State3', 'State4', 'State5', 'State6', 'State7', 'State8', 'State9', 'State10'],
    'Literacy_Rate': [85, 78, 92, 74, 88, 81, 69, 95, 76, 90],
    'Violent_Crimes': [500, 300, 700, 200, 600, 400, 100, 800, 50, 900],
    'Property_Crimes': [2000, 1500, 2500, 800, 2200, 1800, 400, 2600, 300, 2800],
    'Cyber_Crimes': [300, 150, 400, 100, 350, 200, 50, 450, 30, 500],
    'Other_Crimes': [1000, 800, 1200, 500, 1100, 900, 200, 1300, 100, 1400]
}

df = pd.DataFrame(data)
print(df)
```

```python

correlations = df.corr()
print(correlations[['Literacy_Rate']])



plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)
sns.scatterplot(x='Literacy_Rate', y='Violent_Crimes', data=df)
plt.title('Literacy Rate vs Violent Crimes')

plt.subplot(2, 2, 2)
sns.scatterplot(x='Literacy_Rate', y='Property_Crimes', data=df)
plt.title('Literacy Rate vs Property Crimes')

plt.subplot(2, 2, 3)
sns.scatterplot(x='Literacy_Rate', y='Cyber_Crimes', data=df)
plt.title('Literacy Rate vs Cyber Crimes')

plt.subplot(2, 2, 4)
sns.scatterplot(x='Literacy_Rate', y='Other_Crimes', data=df)
plt.title('Literacy Rate vs Other Crimes')

plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 6))
sns.heatmap(correlations, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python

data = {
    'State/UT': ['State1', 'State1', 'State2', 'State2', 'State3', 'State3'],
    'Year': [2020, 2021, 2020, 2021, 2020, 2021],
    'Total_Crimes': [5000, 5200, 3000, 3100, 7000, 6800]
}

df = pd.DataFrame(data)
print(df)
```

```python

print(df.groupby('Year')['Total_Crimes'].sum())


plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Total_Crimes', hue='State/UT', data=df, marker='o')
plt.title('Year-on-Year Total Crime Rate')
plt.xlabel('Year')
plt.ylabel('Total Crimes')
plt.legend(title='State/UT')
plt.show()
```

```python

data = {
    'State/UT': ['State1', 'State2', 'State3', 'State4', 'State5'],
    'Area': [50000, 30000, 70000, 20000, 60000],
    'Total_Crimes': [5000, 3000, 7000, 2000, 6000]
}

df = pd.DataFrame(data)
print(df)
```

```python

correlation = df['Area'].corr(df['Total_Crimes'])
print(f"Correlation between Area and Total Crimes: {correlation}")


plt.figure(figsize=(10, 6))
sns.scatterplot(x='Area', y='Total_Crimes', data=df)
plt.title('Area vs Total Crimes')
plt.xlabel('Area (sq km)')
plt.ylabel('Total Crimes')
plt.show()
```

```python

data = {
    'State/UT': ['State1', 'State2', 'State3', 'State4', 'State5'],
    'Population': [5e6, 3e6, 7e6, 2e6, 6e6],
    'Total_Crimes': [5000, 3000, 7000, 2000, 6000]
}

df = pd.DataFrame(data)
print(df)
```

```python

correlation = df['Population'].corr(df['Total_Crimes'])
print(f"Correlation between Population and Total Crimes: {correlation}")


plt.figure(figsize=(10, 6))
sns.scatterplot(x='Population', y='Total_Crimes', data=df)
plt.title('Population vs Total Crimes')
plt.xlabel('Population')
plt.ylabel('Total Crimes')
plt.show()
```
